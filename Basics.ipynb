```python
import numpy as np
import pandas as pd
from collections import Counter
from IPython.display import display
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import make_classification
from sklearn.metrics import classification_report
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split


data=pd.read_csv('pokemon.csv')
df=pd.DataFrame(data)
display(data.head())
count=df[['Type 1']].groupby('Type 1').size().sort_values(ascending=False)
print(count)

print('''\nJak widać powyżej - największą liczebność mają klasy Water i Normal, takie więc zostaną wybrane do dalszej analizy.\n''')
classes=df[(df['Type 1']=='Water')|(df['Type 1']=='Normal')]
df_limited=classes[['HP','Attack','Defense','Speed','Type 1']]

display(df_limited)

X=np.array(df_limited.iloc[:,0:4])
y=np.array(df_limited['Type 1'])
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2,random_state=4)

neigh = KNeighborsClassifier(n_neighbors=8)
neigh.fit(X_train, y_train)
y_pred=neigh.predict(X_test)

print("\n\033[1mMetoda K najbliższych sąsiadów:\033[0m")
print("\nAccuracy: ",accuracy_score(y_test,y_pred))
print("\nConfusion matrix: \n",confusion_matrix(y_test,y_pred))
print("\nClassification report: \n",classification_report(y_test,y_pred))

tree=DecisionTreeClassifier(random_state=4)
tree.fit(X_train,y_train)
y1_pred=tree.predict(X_test)

print("\n\033[1mMetoda drzewa decyzyjnego:\033[0m")
print("\nAccuracy: ",accuracy_score(y_test,y1_pred))
print("\nConfusion matrix: \n",confusion_matrix(y_test,y1_pred))
print("\nClassification report: \n",classification_report(y_test,y1_pred))

forest = RandomForestClassifier(n_estimators=100,random_state=4)
forest.fit(X_train,y_train)         
y11_pred=forest.predict(X_test)

print("\n\033[1mMetoda Random Forest:\033[0m")
print("\nAccuracy: ",accuracy_score(y_test,y11_pred))
print("\nConfusion matrix: \n",confusion_matrix(y_test,y11_pred))
print("\nClassification report: \n",classification_report(y_test,y11_pred))

log_regr=LogisticRegression(solver='lbfgs')
log_regr.fit(X_train,y_train)
y12_pred=log_regr.predict(X_test)

print("\n\033[1mMetoda regresji logistycznej:\033[0m")
print("\nAccuracy: ",accuracy_score(y_test,y12_pred))
print("\nConfusion matrix: \n",confusion_matrix(y_test,y12_pred))
print("\nClassification report: \n",classification_report(y_test,y12_pred))
```

### Która z metod jest najbardziej skuteczna?
Najbardziej skuteczną metodą jest Random Forest (po polsku: metoda lasów losowych). Jak widać powyżej - posiada ona najwyższą wartość dokładności (accuracy) obliczaną według wzoru: Accuracy = (TP+TN)/(P+N) = (13+16)/((13+6)+(16+7)) = 0.6904761904761905, gdzie P=TP+FN, N=TN+FP. Ma również najwyższe parametry precision (poprawne rozpoznanie w zakresie danej klasy), recall (rozpoznanie ilości elementów danej klasy), f1-score (miara dokładności testu w obrębie klas). Tak samo dla wyliczonych średnich dla tych parametrów - są one najwyższe ze wszystkich metod. W związku z tym wnioskuje się, że metoda Random Forest faktycznie jest najskuteczniejsza.
### Krótka idea działania metody lasów losowych
Random forest jest zespołową metodą uczenia maszynowego, polegającą na konstruowaniu wielu drzew decyzyjnych w czasie uczenia (każde z drzew budowane na innym, losowo wybranym pozbiorze zbioru danych uczących). Decyzja podejmowana "przez głosowanie" - tzn. dla danego zestawu cech wygrywa klasa, do której prowadziło najwięcej drzew. Lasy losowe poprawiają tendencję pojedynczych drzew decyzyjnych do nadmiernego dopasowywania się do danych treningowych, przeuczenia (ang. overfitting), aczkolwiek nie eliminują jej całkowicie.
